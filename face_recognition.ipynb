```python
%matplotlib inline
from matplotlib import pyplot as plt
import cv2

plt.rcParams["figure.figsize"] = (18, 12)
```

## 2. Load cascade, ảnh và chuyển sang ảnh xám

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
img = cv2.imread('_TAO0223.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# cv2 đọc ảnh dạng BGR, matplotlib hiển thị dạng RGB nên
# chuyển đôỉ BGR2RGB để hiển thị được chính xác màu
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
```

## 3. Nhận dạng và vẽ rectangle

```python
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.15, minNeighbors=15)

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 10)

plt.imshow(img)
plt.show()
```

```python
faceId = 1
for (x, y, w, h) in faces:
    fileName = "face."+str(faceId)+ ".jpg"
    cv2.imwrite(fileName, img[y:y+h, x:x+w])
    faceId += 1
    print(f"Write: {fileName}")
```

```python
from IPython.core.display import display, HTML

[display(HTML(f'<img src="face.{i}.jpg" width="100" />')) for i in range(1, faceId)] 
```
